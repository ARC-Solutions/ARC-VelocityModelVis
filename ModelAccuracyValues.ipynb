```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
from matplotlib import font_manager
from matplotlib.font_manager import FontProperties
import matplotlib.patheffects as pe
import re
```

```python
plot_width = 10
plot_height = 6
plot_size = [plot_width, plot_height]
title_font_size = int(3 * plot_height)
axis_label_font_size = int(2.5 * plot_height)
tick_label_font_size = int(2 * plot_height)
autolayout = True
plot_title = "Model Accuracy"
logo_path = r"D:\priv\Programming\ARCV2\backend\CNN\ARC-TextLogo.png"
logo_img = Image.open(logo_path)
title_font = font_manager.FontProperties(fname=r'C:\Users\Ricky\AppData\Local\Microsoft\Windows\Fonts\Formula1-Black.ttf')
label_font = font_manager.FontProperties(fname=r'C:\Users\Ricky\AppData\Local\Microsoft\Windows\Fonts\Formula1-Regular.ttf')
def add_logo():
    # Define the new width and height for the logo
    new_width = int(233 * (fig.dpi / 100))  # scale width by DPI and convert to int
    new_height = int(new_width // 2.31021897810219)  # scale height by DPI

    # Resize the logo
    logo_resized = logo_img.resize((new_width, new_height))

    # Convert the image to RGBA if it's not already
    if logo_resized.mode != 'RGBA':
        logo_resized = logo_resized.convert('RGBA')

    logo_resized_np = np.array(logo_resized)
    logo_x = int((plot_width * fig.dpi - new_width) / 2)  # scale x position by DPI and convert to int
    logo_y = int(55 * (fig.dpi / 100))  # scale y position by DPI and convert to int
    fig.figimage(logo_resized_np, xo=logo_x, yo=logo_y+20, zorder=1, origin='upper', alpha=0.4)
plot_filename = re.sub(r'\W+', '', plot_title.replace("\n", " ")) + ".png"
```

```python
# Use a different style, 'ggplot' is quite modern looking, but there are others like 'seaborn'
plt.style.use('ggplot')

# You can still use dark background if you like it
# plt.rcParams['axes.facecolor'] = 'black'
plt.style.use('dark_background')
print(plt.style.available)

# Increase the size of the plot
fig, ax = plt.subplots(figsize=plot_size, dpi=300)

# Set title parameters
plt.suptitle(plot_title, fontproperties=title_font, size=title_font_size, color='white')

# Plot data with a modern look
plt.plot(history.history['accuracy'], marker='o', markersize=5, linewidth=2, color='c', label='Training accuracy')
plt.plot(history.history['val_accuracy'], marker='o', markersize=5, linewidth=2, color='red', label='Validation accuracy')

# Set labels with larger font size
plt.ylabel('Accuracy', fontsize = axis_label_font_size, fontproperties=label_font, color='white')
plt.xlabel('Epoch', fontsize = axis_label_font_size, fontproperties=label_font, color='white')

# Increase the size of tick labels and change their color
plt.xticks(fontsize=tick_label_font_size, color='white', fontproperties=label_font)
plt.yticks(fontsize=tick_label_font_size, color='white', fontproperties=label_font)

for i, values in enumerate(history.history['accuracy']):
    val = history.history['accuracy'][i]
    text = f"{val: .2f}"
    annotation = plt.annotate(text, (i-0.95,val+0.013), fontsize=10, fontproperties=label_font, color='c')
    annotation.set_path_effects([pe.withStroke(linewidth=1.5, foreground='w')])
for i, values in enumerate(history.history['val_accuracy']):
    val = history.history['val_accuracy'][i]
    text = f"{val: .2f}"
    annotation = plt.annotate(text, (i-0.95,val+0.013), fontsize=10, fontproperties=label_font, color='red')
    annotation.set_path_effects([pe.withStroke(linewidth=1.5, foreground='w')])

# Enable legend and grid
legend = plt.legend(loc='best', prop=label_font)
for text in legend.get_texts():
    text.set_color('white')
plt.grid(True, linestyle='--', alpha=0.6)

add_logo()
plt.tight_layout()
plt.savefig(plot_filename, dpi=fig.dpi)
plt.show()
```
